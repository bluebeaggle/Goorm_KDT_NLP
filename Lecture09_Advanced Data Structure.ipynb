# Stack

- 스택 구조는 기존 List를 활용
    - 동적 배열이기 때문에 push와 pop이 O(1)

```python
a = [1,10]

a.append(4)
a.append(20)

print(a.pop())
print(a.pop())
```

# Queue
- 큐 구조를 기존 List로 만들 경우
 - list는 한 쪽 방향으로 열려 있는 동적 배열
    - 처음 위치 삽입 혹은 삭제가 O(N) 걸림
 - 다른 형태의 데이터 구조가 필요

```python
a= [1,10]

a.insert(0,20)
a.insert(0,15)

print(a)
print(a.pop())
print(a.pop())
print(a)
```

Linked List

- Queue 구조 구현을 위해선 연결 리스트 자료 구조가 필요
 - 양쪽으로 자유로운 입출력
 - 중간 참조는 오래걸리나, 큐 구조에선 상관없음

#***파이썬 collection Library의 deque를 사용!***

```python
from collections import deque               # deque import
queue = deque([10,5,12])                    # deque 생성

queue.appendleft(16)                        # 왼쪽 삽입     O(1)
queue.pop()                                 # 오른쪽 삭제   O(1)
queue.append(20)                             # 오른쪽 삽입   O(1)
queue.popleft()                             # 왼쪽 삭제     O(1)

print(queue)
print(deque(reversed(queue)))               # deque 뒤집기  O(N)
```

#Priority Queue

- 최소/최대값을 빠르게 구할 수 있을까?
 - min/max 함수 사용 -> O(N)
 - 이진 검색을 위해선 내부가 정렬되어 있어야함

- 내부가 항상 정렬된 자료구조를 사용하고 싶다면?
 - 입력할 때 마다 sorted 사용 -> 값 삽입/삭제 : O(N log N)
 - 리스트 정렬 후 값을 중간에 삽입/삭제 -> O(N)

#***파이썬 heqpq Library의 heapq를 사용!***

- 리스트를 이진 트리 형식으로 사용 -> 갑 삽입/삭제: O(log N)

```python
import heapq
queue = [5,2,8,4]                
heapq.heapify(queue)                                    # Heap 초기화, O (N log N)
print(queue)                                            # Heap[k] <= heap[2*k+1], heap[2*k+2]

print(queue[0])                                         # Heap top : 가장 작은 값 위치, Min Heap

heapq.heappush(queue,3)                                 # Heap Push, O(log N)
heapq.heappush(queue, 6)                        
print(queue)

item = heapq.heappop(queue)                             # Heap Pop, O(logN)
print(item, queue[0])

item = heapq.heappushpop(queue, 7)                         # Push하고 Pop 하는 것보다 빠름
print(item, queue[0])
```

# Counter

- 무엇을 세는데 최적화되어 있음
- 집합 연산 지원

```python
from collections import Counter

c = Counter({"korea":2, "English":3})
print(c)
print(c.keys())
print(list(c.elements()))               # 모든 요소 반환
```

```python
from collections import Counter
a= Counter([1,1,2,2,2,3])
b = Counter([2,3,3,4])

print(a+b)              # 횟수 더하기
print(a&b)              # 교집합
print(a|b)              # 합집합
print(a-b)              # 차집합
```
