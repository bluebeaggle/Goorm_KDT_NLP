#**List**

- 배열:일련의 데이터를 하나로 묶음
- 파이썬 배열의 특징
 - 대괄호로 선언 "[value1,value2..]"
 - 아무 타입이나 넣기 가능
 - 길이가 정해져 있지않음

**List Index & Slicing**

- indexing
 - seq[index] 형태로 요소 하나 접근
 - 음수 가능 (뒤에서 접근 시작)


- Slicing
 - seq[start: end: step] 형태로 list 자르기
 - end는 포함되지 않음

**List Operators**
- list + list => list 합치기
- seq[0] = 'something' =>리스트 내에 값 바꾸기
- seq*2  =>곱하기로 여러 개를 여러 번 합치기
- 'something' in seq    => 리스트 안에 요소가 있는지 확인

- len() 함수   => list 길이 구하기
- seq.append()    => 맨 뒤에 요소 추가
- seq.extend([5,6])  => 맨 뒤에 리스트 추가, seq +=[5,6]
- seq.insert(1,1.5)  => 원하는 곳에 삽입 (index, value)

- del seq[1]                => 1번째 요소 삭제, 예약어 형태
- seq.remove('something')   => 원하는 값 삭제

#**Tuple**

- 불변 타입 리스트(Immutable List)
- 선언 시 "[]"가 아닌 "()"을 사용하며, 생략 가능
- 리스트의 연산, 인덱싱, 슬라이싱 등을 동일하게 사용

- 일반적으로 함수에서 2개 이상 요소를 반환할 때 사용
- Tuple은 불변 타입이지만, Tuple안의 요소는 가변타입일 수 있음
- str은 Tuple 형태라고 생각 가능
 - Indexing 및 slicing 가능

#**Packing & Unpakcing**

- 패킹 : 여러 데이터를 묶기
- 언패킹 : 묶인 데이터를 풀기

"_"는 관습적으로 사용하지않는 변수에 사용

```python
t = [1,2,3,4,5]
a,b,c,_,_=t
c
```

- '*(Asterik)'로 남는 부분을 리스트로 남기기 가능

```python
a,*b,c=t
a,b,c
```

#**Dictionary**

- 일종의 매핑을 위한 데이터 구조
 - {}로 생성
 - Key-> Value형태로 구현
 - 불변 타입으로만 이루어져 있으면 Key로 사용 가능

```python
dictionary = {}
dictionary['text'] = 1    #요소 삽입
dictionary['list']=[5,6,7]
dictionary['text'] = 'oh'   #Key값 중복 불가능 -> 덮어씌어짐
```

```python
del dictionary['text']    #요소 삭제
len(dictionary)
```

```python
dictionary = {'한국어':0, '영어': 1, '중국어':2}
```

```python
dictionary.items()
```

```python
dictionary.keys()
```

```python
dictionary.values()
```

```python
2 in dictionary.values()
```

```python
2 in dictionary   #Key값을 기준으로 순회됨
```

#**Set**

- Dictionary의 Key만 모여 있는 형태 -> 집합형

```python
s = set([1,2,3,'text'])
s
```

```python
s.add(4)    #요소 추가
s.add('text') #중복된 요소 추가 불가
s
```

```python
s.remove(2)   #요소 삭제
s.remove('add')   #존재하지 않는 요소는 에러 발생
```

```python
s.discard(99)     #요소 삭제, 존재하지 않은 경우 무시
```

```python
s.update([1,909,None,True])     #여러 요소 추가, 중복은 무시
s
```

```python
s.clear()
s
```

- 수학적 집합 연산자가 존재 
 - '&' 교집합
 - '|' 합집합
 - '-' 차집합
 - '^' 배타적 합집합
