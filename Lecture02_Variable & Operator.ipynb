**원시자료형(Primitive Data Type)들은 불변 타입(Immutable Ype)이다**
- 파이썬의 모든 것은 객체이기 때문에 원시자료형 역시 객체
- 그러나 불변타입들은 저장된 값이 변하지 않는다!
- 모든 타입은 물리적 메모리 주소를 가르킴
- 원시자료형과 Tuple을 제외한 거의 모든 파이썬 객체는 가변타입이다

- 파이썬 대입은 원칙적으로 메모리 주소 복사(즉, 값을 복사하지않고 값을 가르킴)
 - 불변형의 경우 수정이 필요한 경우에 새로운 객체 생성(Out-place 연산)
